```python
def gradient_descent(x, y):
    m_current = b_current = 0
    iterations = 10000
    n = len(x)
    learning_rate = 0.002
    
    previous_cost = 0
    
    # iterate to desuce the low cost value to get best m_current and b_current value
    for i in range(iterations):
        y_predicted = m_current * x + b_current
        pd_m = -(2/n) * sum(x * (y - y_predicted))
        pd_b = -(2/n) * sum((y - y_predicted))
        
        # update m_current and b_current values
        m_current = m_current - learning_rate * pd_m
        b_current = b_current - learning_rate * pd_b
        
        # calculate the cost function
        cost = (1/n) * sum([val**2 for val in (y - y_predicted)])
        
        # stop the iteration of current cost is close to previous cost value
        if math.isclose(cost, previous_cost, rel_tol = 1e-20):
            break
        
        previous_cost = cost
        
        # print the result
        print(f'm: { m_current } | b: { b_current } | cost: { cost } | iteration: { i }')
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

gradient_descent(x, y)
```
